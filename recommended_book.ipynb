```python
from fuzzywuzzy import process
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
book_users = df.pivot(index='book_idN', columns='user_id', values='rating').fillna(0)
matrix_book_users = csr_matrix(book_users.values)
print(matrix_book_users)
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10)
```

```python
knn.fit(matrix_book_users)
```

```python
def recommender(movie_name, data, model, n_recommendations):
    model.fit(data)
    idx = process.extractOne(book_name, df['original_title'])[2]
    print('book Selected:-', df['original_title'][idx], 'Index: ', idx)
    print('Searching for Recommendations...')
    distances, indices = model.kneighbors(data[idx],n_neighbors=n_recommendations)
    for i in indices:
        print(df['original_title'][i].where(i!=idx))
```

```python
recommender('Bayou Moon',matrix_book_users, knn, 5)
```
